```python
import pandas as pd
df = pd.read_csv("Indeks Standar Pencemar Udara di Provinsi DKI Jakarta 2021.csv")
df.tail(35)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['categori'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
df['categori'] = Le.fit_transform(df['categori'])
df
```

```python
df['categori'].value_counts()
```

```python
df['pm25'] = df['pm25'].fillna(df['pm25'].mean())
df.tail(35)
```

```python
new_df = df[['pm10', 'pm25', 'so2', 'co', 'o3', 'no2', 'categori']]
```

```python
new_df.tail(35)
```

```python
from sklearn.model_selection import train_test_split

features = new_df.iloc[:,:6]
label = new_df.iloc[:,6]

x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.30, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model = model.fit(x_train, y_train)
```

```python
print(model.predict(x_test))
print(y_test)
```

```python
from sklearn.metrics import accuracy_score
hasil_prediksi = model.predict(x_test)
accuracy_score(y_test, hasil_prediksi)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(hasil_prediksi, y_test)
```
